```python
import pandas as pd
```

### reading CSV

```python
df = pd.read_csv('data/worldometer_data.csv')
df.head() #it prints the first 5 rows of the data
```

```python
df.columns # it prints the column names in order
```

```python
len(df) # print the total rows
```

### Renaming some columns for preference

```python
df.rename(columns = {
    'Country/Region': 'country',
    'TotalCases': 'confirmed',
    'TotalRecovered': 'recovered',
    'ActiveCases': 'active',
    'TotalDeaths': 'deaths'
}, inplace = True)
```

```python
df.columns
```

### Choosing only the columns in which we are interested

```python
df = df[
    ['country', 'confirmed', 'active', 'deaths', 'recovered']
]
```

### Dropping the NA values rows

```python
df.dropna(inplace = True)
```

```python
len(df)
```

```python
df.sort_values(by = 'confirmed', ascending = False)
```

```python
top_df = df[1:10]
```

```python
top_df
```

### To visualise assignign the columns to the co-ordinated

```python
x = top_df.country
y1 = top_df.confirmed
y2 = top_df.active
y3 = top_df.deaths
y4 = top_df.recovered
```

### Importing matplotlib.pyplot for visualisation

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(x, y1, color = 'blue')
plt.xticks(rotation = 90)
plt.title("Country and Confirmed cases") # Title of the graph
plt.xlabel('Country') # naming to x axis
plt.ylabel('No. of confirmed cases') # naming to x axis
plt.grid()
plt.show()
```

```python
plt.bar(x, y1, color = 'green')
plt.xticks(rotation = 90)
plt.title("Country and Confirmed cases") # Title of the graph
plt.xlabel('Country') # naming to x axis
plt.ylabel('No. of confirmed cases') # naming to x axis
plt.grid()
plt.show()
```

```python
plt.plot(x, y1, label = 'Confirmed')
plt.plot(x, y2, label = 'Active')
plt.plot(x, y3, label = 'Deaths')
plt.plot(x, y4, label = 'Recovered')
plt.xticks(rotation = 90)
plt.title("Country and Confirmed cases") # Title of the graph
plt.legend(loc = 'upper right') # shows the colors used for all different columns
plt.show()
```

### ploting all the graphs in a subplot

```python
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(x, y1, color = 'red')
ax[0, 1].plot(x, y2, color = 'blue')
ax[1, 0].plot(x, y3, color = 'orange')
ax[1, 1].plot(x, y4, color = 'green')
plt.show()
```

```python
df = pd.DataFrame([y1, y2, y3, y4])
df.plot.bar()
plt.legend(x, loc = 'best')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

### Stacking plot

```python
df = pd.DataFrame([y1, y2, y3, y4])
df.plot.bar(stacked = True)
plt.legend(x, loc = 'best')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

### Horizontal bar chart

```python
df.plot.barh()
plt.legend(x, loc = 'best')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

### Area chart

```python
df.plot.area()
plt.legend(x, loc = 'best')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

### Pie chart

```python
plt.pie(y1, labels = x)
plt.title('Confirmed cases')
plt.show()
```
